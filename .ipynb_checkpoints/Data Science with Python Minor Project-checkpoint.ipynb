# Function to evaluate the performance of a classification model

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score

def eval_model(model, x_train, x_test, y_train, y_test):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    
    train_acc = model.score(x_train, y_train)
    print('Train Accuracy:', train_acc)
    
    test_acc = model.score(x_test, y_test)
    print('Test Accuracy:', test_acc)
    
    cm = confusion_matrix(y_test, y_pred)
    cls_rep = classification_report(y_test, y_pred)
    
    print('Confusion Matrix:\n', cm)
    print('Classification Report:\n', cls_rep)

    # New: return metrics for comparison
    return {
        'model': model.__class__.__name__,
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred, average='weighted'),
        'recall': recall_score(y_test, y_pred, average='weighted'),
        'f1_score': f1_score(y_test, y_pred, average='weighted'),
        'model_object': model
    }
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

# Logistic Regression

```python
lr = LogisticRegression(max_iter=10000)
eval_model(lr, x_train, x_test, y_train, y_test)
```

# Decision Tree Classifier

```python
dt = DecisionTreeClassifier(criterion='entropy', max_depth=7, min_samples_split=10)
eval_model(dt, x_train, x_test, y_train, y_test)
```

# Random Forest Classifier

```python
rf = RandomForestClassifier(n_estimators=100, max_depth=7, 
                            criterion='entropy',min_samples_split=10)
eval_model(rf, x_train, x_test, y_train, y_test)
```

### We use F1-score as the primary metric for model selection because it balances both Precision and Recall.
### This is especially useful in cases of class imbalance, where relying on accuracy alone may give misleading results.
### F1-score helps ensure that the model is not only accurate but also reliable in identifying both positive and negative classes correctly.

```python
# Model selection function
def select_best_model(models, x_train, x_test, y_train, y_test, sort_by='f1_score'):
    results = []
    
    for model in models:
        metrics = eval_model(model, x_train, x_test, y_train, y_test)
        results.append(metrics)
    
    sorted_results = sorted(results, key=lambda x: x[sort_by], reverse=True)

    print(f"\nBest Model based on {sort_by.upper()}: {sorted_results[0]['model']}")
    return sorted_results
```

```python
models = [lr, dt, rf]

# Replace with your actual data
results = select_best_model(models, x_train, x_test, y_train, y_test, sort_by='f1_score')
```
